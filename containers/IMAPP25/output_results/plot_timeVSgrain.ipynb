```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("results.csv")
data
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(data["GrainSize"], data["Time[s]"], marker="o", linestyle='-', color='blue', markersize=6, linewidth=1)
plt.title("Execution Time VS Grain Size")
plt.xlabel("Grain Size")
plt.ylabel("Time [s]")
plt.grid(True, alpha=0.7)

plt.subplot(1,2,2)
plt.plot(data["GrainSize"], data["NumTasks"], marker="o", linestyle="-", color="red", markersize=6, linewidth=1)
plt.title("Number of Tasks VS Grain Size")
plt.xlabel("Grain Size")
plt.ylabel("Number of Tasks")
plt.grid(True, alpha=0.7)

plt.tight_layout()
plt.savefig("timevsgrain_log.png", dpi=120)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.plot(data["GrainSize"], data["Time[s]"], marker="o", linestyle='-', color='blue', markersize=6, linewidth=1)
plt.title("Execution Time VS Grain Size (log scale)")
plt.xlabel("Grain Size")
plt.ylabel("Time [s]")
plt.xscale("log")
plt.yscale("log")
plt.grid(True, alpha=0.7)

plt.subplot(1,2,2)
plt.plot(data["GrainSize"], data["NumTasks"], marker="o", linestyle="-", color="red", markersize=6, linewidth=1)
plt.title("Number of Tasks VS Grain Size (log scale)")
plt.xscale("log")
plt.yscale("log")
plt.xlabel("Grain Size")
plt.ylabel("Number of Tasks")
plt.grid(True, alpha=0.7)

plt.tight_layout()
plt.savefig("timevsgrain_log.png", dpi=120)
```
